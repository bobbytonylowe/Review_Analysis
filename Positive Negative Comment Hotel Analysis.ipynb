```python
# neaten up a table and return the values as a transposed table
def Transpose_table(x,name):
    x = x.rename(columns=lambda x: x.replace('Pos_', ''))
    x = x.rename(columns=lambda x: x.replace('Neg_', ''))
    x = x.rename(columns=lambda x: x.replace('neg_', ''))
    x = x.rename(columns=lambda x: x.replace('%', ''))
    x = x.T
    x = x.reset_index()
    x.rename(columns={'index': 'Category'}, inplace=True)
    x.rename(columns={0: name}, inplace=True)
    x['Category'] = x['Category'].str.capitalize()
    x['Category'] = x['Category'].replace('Plates_cups', 'Plates & Cups').replace('Coffee_tea', 
                                                                            'Coffee & Tea').replace('Diet', 'Vegetarian & Vegan').replace('%', '').replace('Loyality', 'Loyalty')
    return x
```

### 2) Create a pivot table looking at the impact of sentiment when someone has mentioned a specific word in the review

- An average score was calculated for the reviews where a category was mentioned in the review
- Here we can analyse the sentiment impact of each word category for each brand
- We can do this by creating a pivot table between the binary field

__________________________________________________________________________________________________________________________________________

```python
#Now lets create a function for the above code

def PT(df, x):
    table= pd.pivot_table(df, index = [x],
                       values = ['sentiment'], aggfunc ={'sentiment':np.mean})
    table = table.reset_index()
    table = table[table[x] == 1]
    table[x] = table[x] * table['sentiment']
    return table[[x]]
```

```python
# Create a function for the above code
def neg_concat_pt_tables(df):
    lst = df.loc[:, 'Neg_Food': 'neg_Technology']
    lst = lst.columns
    table = pd.concat([PT(df, x) for x in lst], axis = 1)
    table = table.rename(columns=lambda x: x.replace('Neg_', ''))
    table = table.rename(columns=lambda x: x.replace('neg_', ''))
    table = table.transpose()
    table = table.reset_index()
    table.rename(columns={'index': 'Category'}, inplace=True)
    table.rename(columns={1: 'Where commented negatively about category'}, inplace=True)
    return table

def pos_concat_pt_tables(df):
    lst = df.loc[:, 'Pos_Food':'Pos_Technology']
    lst = lst.columns
    table = pd.concat([PT(df, x) for x in lst], axis = 1)
    table = table.rename(columns=lambda x: x.replace('Pos_', ''))
    table = table.transpose()
    table = table.reset_index()
    table.rename(columns={'index': 'Category'}, inplace=True)
    table.rename(columns={1: 'Where commented positively about category'}, inplace=True)
    return table
```

```python
def concat_pt_tables(df):
    
    #Merge above dataframes
    x = pd.merge(neg_concat_pt_tables(df) ,pos_concat_pt_tables(df), on=['Category'], how='outer')
    
    # neaten up category values
    x['Category'] = x['Category'].str.capitalize()
    x['Category'] = x['Category'].replace('Plates_cups', 'Plates & Cups').replace('Coffee_tea', 'Coffee & Tea').replace('Diet', 'Vegetarian & Vegan').replace('Loyality', 'Loyalty')
    
    # problems with just subtracting columns - therefore used pandas diff() function and renamed the columns and joined
    test = x.diff(axis=1)
    test.rename(columns={'Where commented positively about category': 'Difference'}, inplace=True)
    
    # Concat difference value to impact table
    test = test[['Difference']]
    concat_difference_table = pd.concat([x,test], axis = 1)
    
    # Merge Impact tabe with positive frequency table
    Positive_and_impact = pd.merge(pos_table(df), concat_difference_table, on=['Category'], how='inner')
    
    # Merge Impact tabe with negative frequency table
    impact_and_freq = pd.merge(neg_table(df), Positive_and_impact, on=['Category'], how='inner')
    
    # Calculate Total Mention
    impact_and_freq['Total Mentions'] = impact_and_freq['# of Reviews Negatively talking about category'] + impact_and_freq['# of Reviews Positively talking about category']
    impact_and_freq = impact_and_freq.drop_duplicates()

    #return table with most mentioned category first
    return impact_and_freq.sort_values('Total Mentions', ascending=False)
```

```python
# function to create an excel file with multiple sheets
def dfs_tabs(df_list, sheet_list, file_name):
    
    '''function to create an excel file with multiple sheets '''
    
    writer = pd.ExcelWriter(file_name,engine='xlsxwriter')   

    for dataframe, sheet in zip(df_list, sheet_list):
        dataframe.to_excel(writer, sheet_name=sheet, startrow=0 , startcol=0)   
        
    writer.save()
```
